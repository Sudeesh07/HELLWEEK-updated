```python
# Categorical columns
cat_col = [col for col in df.columns if df[col].dtype == 'object']
print('Categorical columns :',cat_col)
# Numerical columns
num_col = [col for col in df.columns if df[col].dtype != 'object']
print('Numerical columns :',num_col)
```

```python
def cleaner(row):

    if row in [np.nan,'nan',0.0]:
        return row
    else:
        x,y=6,-6
        try:
            return float(row[x:y])
        except:
            print(f'{row, type(row)}, exception')
def pre_process(df, mode):
    df = df.drop(['UID', 'ph_no', 'cvv', 'credit_card_number', 'job', 'email', 'url', 'country', 'emoji', 'name'], axis=1)
    n = len(df.columns) if mode == 'test' else -1
    for i in df.columns[:n]:
        df[i] = df[i].apply(cleaner)
    df = df.fillna(df.median())
    return df
df = pre_process(df, 'train')
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns
df.select_dtypes(include=['object']).columns
```

```python
from sklearn.preprocessing import LabelEncoder as LE
x=df.drop('state',axis=1)
y=df['state']
le=LE()
y=le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
import pandas as pd
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
X_train
len(X_train),len(y_train),len(X_test),len(y_test)
X_train
```

```python
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
from sklearn.metrics import accuracy_score
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout

model = Sequential()

model.add(Dense(256, activation='relu', input_shape=(X_train.shape[1],)))

model.add(Dense(128, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.3))

model.add(Dense(y_train.shape[1], activation='softmax'))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=50, batch_size=128, verbose=1)
```

```python
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)  # Convert probabilities to class labels
accuracy = accuracy_score(np.argmax(y_test, axis=1), y_pred_classes)
print(f"Accuracy: {accuracy}")
```
